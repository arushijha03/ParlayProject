```python
import pandas as pd
```

```python
ind_det=pd.read_csv('/content/ind_det.csv')
ind_det_credit_debts= pd.read_csv('/content/ind_det_credit_debts.csv')
ind_det_debts =pd.read_csv('/content/ind_det_debts.csv')
ind_det_expenses = pd.read_csv('/content/ind_det_expenses.csv')
ind_det_incomes= pd.read_csv('/content/ind_det_incomes.csv')
```

```python
credit_debts_n_debts = ind_det_credit_debts.merge(ind_det_debts)
```

```python
credit_debts_n_debts.head()
```

```python
credit_debts_n_debts.info()
```

```python
expenses_n_incomes = ind_det_expenses.merge(ind_det_incomes)
```

```python
expenses_n_incomes.head()
```

```python
expenses_n_incomes.info()
```

```python
credit_debts_n_debts_n_expenses_n_incomes = credit_debts_n_debts.merge(expenses_n_incomes)
```

```python
credit_debts_n_debts_n_expenses_n_incomes.head()
```

```python
credit_debts_n_debts_n_expenses_n_incomes.info()
```

```python
credit_debts_n_debts_n_expenses_n_incomes_n_ind_det = credit_debts_n_debts_n_expenses_n_incomes.merge(ind_det)
```

```python
credit_debts_n_debts_n_expenses_n_incomes_n_ind_det.head()
```

```python
credit_debts_n_debts_n_expenses_n_incomes_n_ind_det.info()
```

```python
credit_debts_n_debts_n_expenses_n_incomes_n_ind_det.to_csv("ind_det_final.csv", index=False)
```
