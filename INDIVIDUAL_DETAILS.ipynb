```python
ind_det_incomes = ind_det.copy()
```

```python
# assume your DataFrame is named ind_det_income and has columns: inquiry_id, incomes
FIELDS = ["id", "name", "type", "amount"]

def as_list_of_dicts(x):
    """Ensure each cell becomes a list[dict]; accept JSON text / dict / list / None."""
    x = json.loads(x) if isinstance(x, str) else x
    if x is None:
        return []
    return x if isinstance(x, list) else [x]

# build the four list-columns from 'incomes'
for f in FIELDS:
    ind_det_incomes[f] = ind_det_incomes["incomes"].apply(
        lambda cell: [d.get(f) for d in as_list_of_dicts(cell)]
    )

# (optional) make 'amount' values numeric inside each list (invalids -> NaN)
ind_det_incomes["amount"] = ind_det_incomes["amount"].apply(
    lambda lst: [pd.to_numeric(v, errors="coerce") for v in lst]
)

# keep ONLY these five columns, with inquiry_id first
ind_det_incomes = ind_det_incomes[["inquiry_id"] + FIELDS].copy()
```

```python
ind_det_incomes = ind_det_incomes.rename(columns={"id": "ind_det_incomes_id", "name": "ind_det_incomes_name","type": "ind_det_incomes_tyoe", "amount": "ind_det_incomes_amount"})
```
